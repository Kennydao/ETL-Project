```python
with open("file.xml") as xml_file:
    data_dict = xmltodict.parse(xml_file.read())
xml_file.close()
```

```python
json_data = json.dumps(data_dict)
```

```python
with open("data.json", "w") as json_file:
    json_file.write(json_data)
```
